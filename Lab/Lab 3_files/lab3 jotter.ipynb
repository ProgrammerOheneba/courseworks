```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
w0 =1  # m/s
a0 = 0  # m/s^2
N_sq = 1e-4  # 1/s^2
t_max = 3600  # s
dt_1 = 1/(N_sq)**0.5  # s
dt_2 = 2/(N_sq)**0.5  # s

# Define exact solution
def w_exact(t):
    return w0*np.cos((N_sq)**0.5*t) + a0/(N_sq)**0.5*np.sin((N_sq)**0.5*t)

# Define numerical solution
def w_numerical(dt):
    # Set initial conditions
    phi0 = w0/(N_sq)**0.5
    phi1 = phi0 - a0*dt/(N_sq)**0.5
    # Iterate over time steps
    phi_list = [phi0, phi1]
    t_list = [0, dt]
    while t_list[-1] < t_max:
        phi_next = 2*phi_list[-1] - phi_list[-2] - N_sq*dt**2*phi_list[-1]
        phi_list.append(phi_next)
        t_list.append(t_list[-1] + dt)
    # Convert phi values to w values
    w_list = [N_sq**0.5*phi for phi in phi_list]
    return np.array(w_list)

# Evaluate solutions
t = np.linspace(0, t_max, 1000)
w_exact_1 = w_exact(t)
w_exact_2 = w_exact(t)
w_numerical_1 = w_numerical(dt_1)
w_numerical_2 = w_numerical(dt_2)

# Plot solutions
fig, axs = plt.subplots(nrows=2, sharex=True, figsize=(8, 8))
axs[0].plot(t, w_exact_1, label='Exact')
axs[0].plot(np.arange(len(w_numerical_1))*dt_1, w_numerical_1, label='Numerical')
axs[0].set_title(r'$N\Delta t=1$')
axs[1].plot(t, w_exact_2, label='Exact')
axs[1].plot(np.arange(len(w_numerical_2))*dt_2, w_numerical_2, label='Numerical')
axs[1].set_title(r'$N\Delta t=2$')
for ax in axs:
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Displacement (m)')
    ax.legend()
plt.tight_layout()
plt.show()
```

$$\dfrac{w((n+1)\Delta t) - 2w(n\Delta t) + w((n-1)\Delta t)}{(\Delta t)^2} + N^2 w(n\Delta t) = -\dfrac{e_{n+1} - 2e_n + e_{n-1}}{(\Delta t)^2} - N^2 e_n.$$
