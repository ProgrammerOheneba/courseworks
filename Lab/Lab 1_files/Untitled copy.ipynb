```python
import calendar

def julian_to_gregorian(year, month, day):
    # Check if the given year is a leap year in the Julian calendar
    if calendar.isleap(year):
        # February 29: Julian leap day
        if month == 2 and day == 29:
            gregorian_date = f"{year}-03-01"
        elif month == 3 and day == 1:
            gregorian_date = f"{year}-03-02"
        else:
            gregorian_date = f"{year}-{month:02}-{day:02}"
    else:
        gregorian_date = f"{year}-{month:02}-{day:02}"

    return gregorian_date

# Sample dates
dates = [
    (121, 1, 16),
    (121, 2, 15),
    (121, 3, 16)
]

gregorian_dates = [julian_to_gregorian(*date) for date in dates]
print(gregorian_dates)
```
